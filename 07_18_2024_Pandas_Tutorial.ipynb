## 7. Data Visualization
### 7.1 Visualize the Distribution of Passengers' Ages

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=10, edgecolor='black', alpha=0.7)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

### 7.2 Visualize the Survival Rate by Class

```python
survival_rate_by_class = df.groupby('Pclass')['Survived'].mean()

plt.figure(figsize=(10, 6))
plt.bar(survival_rate_by_class.index, survival_rate_by_class.values, color='skyblue')
plt.title('Survival Rate by Class')
plt.xlabel('Class')
plt.ylabel('Survival Rate')
plt.xticks(rotation=0)
plt.show()
```

## 7.3 Data Corelation
#### A heatmap of the correlation coefficients between the columns 'Age', 'Fare', 'Survived', and 'Pclass'

```python
import seaborn as sns
correlation_matrix_sample = df[['Age', 'Fare', 'Survived','Pclass']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_sample, annot=True, fmt=".2f", cmap='coolwarm', cbar_kws={'label': 'Correlation coefficient'})
plt.title('Correlation Matrix for Sample Titanic Dataset')
plt.show()
```

```python
df = pd.read_excel('Math_Grades_Dataset.xlsx')
df.set_index('Student ID', inplace=True)
df
```

#### 8.1 Line Plot

```python
plt.figure(figsize=(10, 6))
plt.plot(df.columns, df.loc['S5'], label='S5')
plt.title('Grades Over Months for Student 5')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.ylabel('Grade')
plt.legend(title='Student ID', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### 8.2. Heatmap

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df, annot=True, fmt="d", cmap='YlGnBu', cbar_kws={'label': 'Grade'})
plt.title('Heatmap of Grades')
plt.xlabel('Month')
plt.ylabel('Student ID')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### 8.3. Boxplot

```python
plt.figure(figsize=(10, 6))
box = plt.boxplot(df.loc['S5'], patch_artist=True, boxprops=dict(facecolor='lightblue', color='blue'),
                  whiskerprops=dict(color='blue'), capprops=dict(color='blue'), medianprops=dict(color='red'))

plt.title('Box Plot of Grades for Student S5')
plt.xlabel('Student S5')
plt.ylabel('Grade')
plt.xticks([1], ['S5'])
plt.tight_layout()
plt.show()
```

# Object Oriented Programming

```python
class Car:
    def __init__(self, color, mileage):
        self.color = color
        self.mileage = mileage

    def describe(self):
        return f"A {self.color} car with {self.mileage} miles."

    def drive(self, miles):
        self.mileage += miles
```
